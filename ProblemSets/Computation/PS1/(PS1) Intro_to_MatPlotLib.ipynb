```python
import numpy as np
import scipy as sy
from matplotlib import pyplot as plt
```

### Problem 1

```python
def varmean(n):
    sample = np.random.normal(size=(n,n))
    mean_by_row = np.mean(sample,1)
    varmean = np.var(mean_by_row)
    return varmean
```

```python
x = 100*np.arange(1,11)
y = [varmean(n) for n in x]
plt.plot(x,y)
plt.xlabel("n = no. of sample")
plt.ylabel("variance of means")
plt.show()
```

### Problem 2

```python
x = np.linspace(-2*np.pi,2*np.pi,100)
sinx = np.sin(x)
cosx = np.cos(x)
arctanx = np.arctan(x)
plt.plot(x, sinx, x, cosx, x, arctanx)
plt.legend(["sin(x)","cos(x)","arctan(x)"],loc='upper left')
plt.show()
```

### Problem 4

```python
x = np.linspace(0, 2*np.pi, 100)

fig, axes = plt.subplots(2, 2, figsize=[10,10])
plt.axis([0, 2*np.pi, -2, 2])

ax1 = plt.subplot(221)
ax1.plot(x, np.sin(x),'g-')
plt.title("sin(x)")
          
ax2 = plt.subplot(222)
ax2.plot(x, np.sin(2*x),'r--')
plt.title("sin(2x)")

ax3 = plt.subplot(223)
ax3.plot(x, 2*np.sin(x),'b--')
plt.title("2sin(x)")

ax4 = plt.subplot(224)
ax4.plot(x, 2*np.sin(2*x),'m:')
plt.title("2sin(2x)")
```
